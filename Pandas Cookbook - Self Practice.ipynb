```python
df3 = pd.DataFrame({'Name':['Nikhil', 'Ankit', 'Keval', 'Darpan', 'Rajesh', 'John', 'Lynda'],
                  'Score':[89,92,96,82,95,98,97], 
                  'Type':['Fat','Slim','Fat','Slim','Fat','Slim','Slim'],
                  'Country':['India','USA','Denmark','Australia','Italy','China','Israel'],
                  'Surname':['Sharma','Sharma','Patel','Shah','Sharma','Sharma','Sharma']}                 
                ).set_index('Country').set_index('Surname', append=True)
```

```python
df3
```

```python
# https://stackoverflow.com/questions/62065179/pandas-multiindex-selection-criteria-alongwith-column-selection-criteria/62065331#62065331

'''Country should neither be India nor be USA
Surname should be Sharma
Score should be equal to or above 90
Type should be Slim'''
```

```python
df3.query('Country not in ["India","USA"] and Surname == "Sharma" and Score >= 90 and Type =="Slim"')
```

```python
# Same as above but without using Query

m1 = ~df3.index.get_level_values('Country').isin(["India","USA"])
m2 = df3.index.get_level_values('Surname') == 'Sharma'
m3 = df3['Score'].ge(90)
m4 = df3['Type'].eq('Slim')

print (df3[m1 & m2 & m3 & m4])
```

```python
scores = pd.read_excel('jumbled_scores.xlsx')
```

```python
scores
```

```python
# Here the data is not tidy. The years should be data, but they exist as columns
```

```python
scores.melt(id_vars=['Name'], var_name=['Year'])
```

```python
# let's sort

scores.melt(id_vars=['Name'], var_name=['Year']).sort_values(['Year', 'Name'])
```

```python
# let's create a pivot table for better readability

scores.melt(id_vars=['Name'], 
            var_name=['Year']) \
            .sort_values(['Year', 'Name']) \
            .pivot_table(index=['Name'], columns=['Year'], values=['value'])
```

```python
# let's remove the 'value' tag that is there above the actual years by using 'droplevel'

scores.melt(id_vars=['Name'], 
            var_name=['Year']) \
            .sort_values(['Year', 'Name']) \
            .pivot_table(index=['Name'], columns=['Year'], values=['value']) \
            .droplevel(0, axis=1)
```
